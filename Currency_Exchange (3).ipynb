```python
import pandas as pd
import numpy as np
```

```python
df=pd.read_csv("/content/drive/MyDrive/Colab Notebooks/Datasets/Currency.csv")
df['date']=pd.to_datetime(df['date'])
df.dtypes
```

```python
df = df[["AED", "USD", "GBP", "INR", "PHP"]]
df
```

```python
from sklearn.preprocessing import MinMaxScaler
scaler = MinMaxScaler()
df = pd.DataFrame(scaler.fit_transform(df), columns=df.columns, index=df.index)
```

```python
xtrain=df[:3000]
xtest=df[3000:]
xtrain[:11]
xtest
```

```python
from keras.preprocessing.sequence import TimeseriesGenerator
xtrain_generator = TimeseriesGenerator(xtrain.values, xtrain.values,length=60, batch_size=1)
xtrain_generator[0]
```

```python
from keras.models import Sequential
from keras.layers import Dense,LSTM,Dropout
from keras.optimizers import Adam
```

```python
model = Sequential()
model.add(LSTM(128, activation='relu', return_sequences=True, input_shape=(60, 5)))
model.add(Dropout(0.2))
model.add(LSTM(128, activation='relu', return_sequences=True))
model.add(Dropout(0.2))
model.add(LSTM(128, activation='relu'))
model.add(Dense(5))
```

```python
optimizer = Adam(learning_rate=0.001)
model.compile(optimizer=optimizer, loss='mse')
```

```python
model.fit(xtrain_generator,epochs=3)
```

```python
test_sample=xtrain.tail(60)
test_sample=test_sample.values
test_sample=test_sample.reshape(1,60,5)
model.predict(test_sample)
test_sample
```

```python
np.append(test_sample[:,1:],[[[.2,1,2,3,3]]],axis=1)
test_sample
```

```python
prediction=[]
for i in range(100):
  pred=model.predict(test_sample)
  prediction.append(np.array(pred))
  test_sample=np.append(test_sample[:,1:],[pred],axis=1)
```

```python
prediction
```

```python
combined_array=np.vstack(prediction)
dfpred=pd.DataFrame(combined_array, columns=["AED", "USD", "GBP", "INR", "PHP"])
dfpred=pd.DataFrame(scaler.inverse_transform(dfpred), columns=dfpred.columns)
tail=pd.DataFrame(scaler.inverse_transform(xtest.tail(100)),columns=["AED", "USD", "GBP", "INR", "PHP"])
print(dfpred)
print(tail)
```

```python
from sklearn.metrics import mean_squared_error,r2_score
lst=[]
for i in range(0,100):

  lst.append(r2_score(tail.values[i], dfpred.values[i]))
print(f"The average R2 of the Model is : {np.mean(lst)*100} %")
```
